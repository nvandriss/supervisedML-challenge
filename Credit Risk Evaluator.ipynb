```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data

YN_dict = {'Y':1,'N':0}
homeowner_dict = {'ANY':0,'RENT':1,'MORTGAGE':2,'OWN':3}
veri_dict = {'Not Verified':0,'Source Verified':1,'Verified':1}
loan_dict = {'low_risk':1,'high_risk':0}
initial_list_dict = {'w':0,'f':1}
apptype_dict = {'Individual':1,'Joint App':0}

train_df2 = train_df.replace({'hardship_flag':YN_dict, 'debt_settlement_flag':YN_dict})
train_df3 = train_df2.replace({'home_ownership':homeowner_dict})
train_df4 = train_df3.replace({'verification_status':veri_dict})
train_df5 = train_df4.replace({'loan_status':loan_dict})
train_df6 = train_df5.replace({'initial_list_status':initial_list_dict})
train_df7 = train_df6.replace({'application_type':apptype_dict})

Train_df = train_df7.drop(['index','Unnamed: 0','pymnt_plan'],axis='columns')

file_path = Path('Resources/cleaned_2019_credit_data.csv')
Train_df.to_csv(file_path, index=False)

Train_df.head()
```

```python
X_train = Train_df.drop('loan_status', axis=1)
y_train = Train_df['loan_status'].values
print(X_train.select_dtypes(include=[object]))
```

```python
# Convert categorical data to numeric and separate target feature for testing data
test_df2 = test_df.replace({'hardship_flag':YN_dict, 'debt_settlement_flag':YN_dict})
test_df3 = test_df2.replace({'home_ownership':homeowner_dict})
test_df4 = test_df3.replace({'verification_status':veri_dict})
test_df5 = test_df4.replace({'loan_status':loan_dict})
test_df6 = test_df5.replace({'initial_list_status':initial_list_dict})
test_df7 = test_df6.replace({'application_type':apptype_dict})

Test_df = test_df7.drop(['index','Unnamed: 0','pymnt_plan'],axis='columns')

file_path = Path('Resources/cleaned_2020_credit_data.csv')
Test_df.to_csv(file_path, index=False)

Test_df.head()
```

```python
# add missing dummy variables to testing set
Test_df = pd.get_dummies(Test_df)
X_test = Test_df.drop('loan_status', axis=1)
y_test = Test_df['loan_status'].values
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',max_iter=200)
classifier.fit(X_train, y_train)
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```
